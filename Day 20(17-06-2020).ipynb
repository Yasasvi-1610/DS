### OneHotEncoder --> similar to pd.get_dummies

--> Import the OneHotEncoder Class <br>
-->Create an object for the class<br>
--> ColumnTransformer-->Insider this columntransform,we will pass this OneHotEncoder<br>
--> It has two mandatory arguments--> transforms,remainder

```python
## Importing librarues
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Insider the columntransformer, we have to pass the OneHotEncoder
column_transformer= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],
                                      remainder='passthrough')
```

```python
#Syntax: objectname.fit_transform(data)
```

```python
data=column_transformer.fit_transform(data)
data
```

```python
X=pd.DataFrame(data)
X
```

### Feature Scaling:(Not mandatory for all the algorithms)

```python
sample=pd.DataFrame({'W.E':[2,4,5,6,7,8],'Age':[28,25,32,45,33,32],
                    'Salary':[300000,400000,500000,600000,700000,800000],
                     'Distance':[10,12,23,32,15,17],
                    'Expected Salary':[400000,500000,600000,700000,800000,900000]})
sample
```

--> Most of the Machine Learning Algorithms, we use distance based formulas

--> Salary column is hugely dominating the other columns in terms of magnitude<br>
-->Also, all the columns in our dataset, belong to different units, Exp--> Years,Salary -->Currency<br>
-->Units and Scale (magnitude)<br>
-->Feature Scaling -->when applied, this units and scaling will be remmoved<br>
--> To make the data unitless and scaless, we have to apply Feature Scaling

```python
from sklearn.preprocessing import scale
scaled_data=scale(sample)
scaled_data
```

```python
scaled_data=pd.DataFrame(scaled_data)
scaled_data
```

```python
standard_data=sample.copy() #creating copy of original dataset
standard_data
```

```python
#Using StandardScaler
from sklearn.preprocessing import StandardScaler
sc= StandardScaler() # creating an object
sample=sc.fit_transform(sample)
sample
```

```python
sample=pd.DataFrame(sample)
sample
```
