```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("a1.jpg")
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
```

```python
plt.imshow(img[:,:,0],cmap='gray')
```

```python
plt.imshow(img[:,0:1000])
```

```python
plt.imshow(img[:400,:1000])
```

```python
plt.imshow(img[600:800,1250:1750])
```

```python
plt.imshow(img[:00,:1000])
```
