```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data= pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv")
data
```

```python
data=data.replace({'temperature': -99999,'windspeed': -99999,'event': '0'},np.nan)
data
```

### interpolate() -based on column values it will come up with a guess values to fill up with NaN value

```python
int_data=data.interpolate()
int_data
```

### dropna(): Drops the columns or rows which have the missing values

```python
drop_data=data.dropna()
drop_data
```

```python
# any meaning finds any missing values it drops a whole row
drop_data=data.dropna(how='any')
drop_data
```

```python
#drop the columns having any missing values
drop_data=data.dropna(how='any',axis=1)
drop_data
```

```python
drop_all_data=data.dropna(how='all')
drop_all_data
```

### Handling Missing Values using Scikit-Learn

```python
### SimpleImputer
```

```python
# creating an object for SimpleImputer class
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
# It is a 2 step transformation,applying the fit method on the data
imputer.fit(data.iloc[:,1:3]) #indices of column1 and column2
```

```python
#Transforming the data
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3])
data
```

### Encoding Categorical Data

#### Categorial Data Types
    - Nominal
    - Ordinal
Using Pandas Library--> get_dummies()

```python
price=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv")
price
```

```python
#Covert the categoricaldata into numerical data
dummy_set=pd.get_dummies(price.town)
dummy_set #dummy columns
```

```python
merged_data=pd.concat([price,dummy_set])#,axis=1)
merged_data
```

```python
merged_data=pd.concat([price,dummy_set],axis=1)
merged_data
```

```python
pd.get_dummies(price,columns=['town'])
```

```python
pd.get_dummies(price,columns=['town'],drop_first=True)
```
