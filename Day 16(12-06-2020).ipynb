```python
#
a=input()
list1=a.split(",")
list2=[i.split(":")[1] for i in list1]
list3=[]
for num in list2:
    n=int(num)
    rev=num[::-1]
    while str(n)!=rev:
        n+=int(rev)
        rev=str(n)[::-1]
    list3.append(n)
list1[list3.index(max(list3))].split(":")[0]
```

```python
def rev_number(n):
  s = 0
  while True:
    k = str(n)
    if k == k[::-1]:
      break
    else:
      m = int(k[::-1])
      n += m
      s += 1
  return n 

print(rev_number(265))
```
