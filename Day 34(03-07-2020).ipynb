```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Reading the data
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv",
                  encoding = 'latin-1')
data.head()
```

```python
data.shape
```

```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1, inplace = True)
data.shape
```

```python
#Split the data
#80% of the data will be in training data and 20% of the data will be in test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.v2, data.v1, test_size= 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#CountVertorizer
from sklearn.feature_extraction.text import CountVectorizer

#Creating an object for CountVertorizer
count_vect = CountVectorizer()
```

```python
#Generate the word counts for the words in the documents 
word_count_vector = count_vect.fit(X_train)
word_count_vector
```

```python
word_count_vector.get_feature_names()
```

```python
doc_array = word_count_vector.transform(X_train)
doc_array
```

```python
#Converting to an array
doc_array = word_count_vector.transform(X_train).toarray()
doc_array
```

```python
#Converting array into dataframe
feature_matrix = pd.DataFrame(doc_array, columns = word_count_vector.get_feature_names())
feature_matrix
```

```python
#pd.set_option('display.max_rows', None)
```

```python
#TFIDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#intialize and object for TFIDF vectorizer
tfidf = TfidfVectorizer()
```

```python
# apply the tfidf to the data(X_train)
X_train_transformed = tfidf.fit_transform(X_train)
X_train_transformed
```

```python
X_test_transformed = tfidf.transform(X_test)
X_test_transformed
```

```python
#Feature Nmaes
tfidf.get_feature_names()
```

```python
#Position of the words in the sparse matrix
tfidf.vocabulary_
```

```python
# Idf of the terms
tfidf.idf_
```

```python
# Apply the Naive Bayes Algorithm 

#Import BernoulliNB
from sklearn.naive_bayes import BernoulliNB

#Creating an object for BerNb
model_BerNB = BernoulliNB()
```

```python
#Applying the algorithm to the data
#ObjectName.fit(Input, Output)

model_BerNB.fit(X_train_transformed, y_train)
```

```python
#Prediction on test data
#Syntax: objectName.predict(InputValues)
y_test_pred = model_BerNB.predict(X_test_transformed)
```

```python
#Compare the actual values(y_test) with predictes values(y_test_pred)
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_test_pred)
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
#Test a singlr message--> Spam or Ham
new_message = pd.Series('Last few days for the to collect the free entry passes with 40% discount')

#Calcuate tfidf values for the new message
new_message_transformed = tfidf.transform(new_message)
```

```python
#Predict the new message
model_BerNB.predict(new_message_transformed)
```
