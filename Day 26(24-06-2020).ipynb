```python
# To create a dataframe out of dictionary
boston_df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
boston_df
```

```python
#Created a new column with name MEDV and passed the target values from the dictionary
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
# We are using RM columns as input(SLR --> we have only one input)
#Output-->MEDV column
```

```python
#Preparing the input and output columns
X = boston_df[['RM']] #selecting column as dataframe
X
```

```python
X = boston_df['MEDV'] #selecting column as a series
X
```

```python
X = boston_df[['RM']] #selecting column as dataframe
y = boston_df['MEDV']
y
```

```python
# plt.style.available
# plt.style.use
with plt.style.context("grayscale"): #temporary styling
    plt.plot(X,y,"gs") #bo- blue circle
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("Scatter Plot")
```

```python
# Usual Sklearn: import, instaniate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,y) #it will try to find optimum m and c ,error should be minimized
```

```python
# Print intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
#y=mx+c calcuating the predictions
#m value--> coefficient(9.10210898)
#c value--> intercept(-34.67062078)
(9.10210898 * 6.575)+( -34.67062078)
```

#### Interpreting the model:
A unit increase in X(RM) is associated with 9.10210898 unit increase in MEDV

```python
# Pipeline --> First we create a model on the data
#Second, we need to compare the model(predicted values with actual values)
```

```python
# With the help of predict function, we are going to find our predicted values
y_pred = lm.predict(X)
y_pred
```

```python
#comparing actual data with predicted data
y==y_pred
```

```python
##r2_value--> to check the model performance
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
# Observe the regression line
with sns.axes_style('dark'):
    sns.regplot(X,y)
```

```python
#Scatterplot(Regreesion)
#- First we are making a scatter plot of the original data(X and Y)
#- We need to plot a line for the predicted values(X,y_pred)
plt.scatter(X,y,color='black')
plt.plot(X,y_pred,color='red')
plt.xlabel('RM I/P')
plt.ylabel('MEDV O/P')
```
