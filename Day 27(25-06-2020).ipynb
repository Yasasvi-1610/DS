```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the boston data from sklearn
from sklearn.datasets import load_boston
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
#Created a new column with name MEDV and passed the target values from the dictionary
boston_df['MEDV'] = boston.target
boston_df.head()
```

### To find the correlation we use corr():

```python
boston_df.corr()
```

### Visiualizing the correlation:

```python
sns.heatmap(boston_df.corr())
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(), annot = True, cmap='RdBu')
```

```python
# Correlation>0.7
correlation = boston_df.corr()
cor = correlation[correlation>0.7]
plt.figure(figsize=(14,14))
sns.heatmap(cor, cmap = 'Greens', annot = True)
```

```python
#correlation>0.7 and correlation<0.9
cor = correlation[(correlation>0.7) & (correlation<0.9)]
plt.figure(figsize=(14,14))
sns.heatmap(cor, cmap = 'Greens', annot = True)
```

```python
#Split the data into training and testing. Built the model on training data and test the model on testing data
```

```python
#Split the data into Input and Output
X = boston_df.drop(['MEDV'],axis=1) #Input columns
y = boston_df.MEDV # boston_df['MEDV']
```

```python
# Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Build the model on Traing data --> X_train and y_train
# sklearn library: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#Intercept and the coefficient values
print(lm.intercept_)
lm.coef_
```

```python
# Create a dataframe for coefficients
coefficients = pd.DataFrame([X_train.columns])
coefficients
```
