```python
#6. Get the row which has highest population
df[df.population_total==max(df['population_total'])]
```

```python
#7. Calculate the mean of the population total grouped by States and sort the population total in ascending order 
df2=df.groupby(['state_name'])['population_total'].mean()
df2=pd.DataFrame(df2)
df2.sort_values(by='population_total')
```

```python
#7. plot a horizontal bar chart for the mean of the population total grouped by States in ascending order
df2.plot.barh(figsize=(5,5))
```

```python
#8. Create a dataframe of five cities which have lowest population
df1=df.sort_values(by='population_total').head()
df1
```

```python
#9. Create a dataframe of  five cities which have highest  population
df2=df.sort_values(by='population_total',ascending=False).head()
df2
```

```python
#10. Concatenate the top five and bottom five cities based on the population
pd.concat([df1,df2])
```
