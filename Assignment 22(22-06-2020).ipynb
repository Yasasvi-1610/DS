```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/cities.csv")
data.head()
```

```python
#a.Analyze the  top 5 cities with highest number of female graduates
df1=data.sort_values(by='female_graduates',ascending=False).head(5)
df1['name_of_city']
```

```python
#b.Find the 5 cities with lowest number of female graduates
#and concatenate the two dataframes of top 5 cities with  bottom 5 cities
df2=data.sort_values(by='female_graduates',ascending=True).head(5)
df2['name_of_city']
```

```python
frames=[df1,df2]
pd.concat(frames)
```

```python
#d.Find the  top 10 cities with highest total graduation rate
df3=data.sort_values(by='total_graduates',ascending=False).head(10)
df3['name_of_city']
```

```python
#e.Calculate the average child sex ratio of the Kerala State.
df4=data.groupby('state_name')['sex_ratio'].mean()
df4['KERALA']
```
